```python
import os
import json
import torch
import numpy as np
import cv2
from ultralytics import YOLO
from google.colab import files
from IPython.display import display, Image

# 1. SETUP MODELS
parts_path = '/content/drive/MyDrive/CarParts_YOLO26/yolo26_carparts_best.pt'
if not os.path.exists(parts_path): parts_path = 'yolo26_carparts_best.pt'

damage_path = '/content/drive/MyDrive/CarParts_YOLO26/damage_detection_best.pt'
if not os.path.exists(damage_path): damage_path = 'damage_detection_best.pt'

model_parts = YOLO(parts_path)
model_damage = YOLO(damage_path)

def extract_and_assess_segments(filename, model_parts, model_damage):
    orig_img = cv2.imread(filename)
    h, w = orig_img.shape[:2]

    res_parts = model_parts.predict(source=orig_img, conf=0.25, verbose=False)[0]

    # --- PHASE 1: Cross-Class Overlap Filtering (NMS) ---
    indices_to_drop = set()
    if res_parts.boxes is not None and len(res_parts.boxes) > 1:
        boxes = res_parts.boxes.xyxy.cpu().numpy()
        confs = res_parts.boxes.conf.cpu().numpy()
        cls_ids = res_parts.boxes.cls.cpu().numpy().astype(int)

        conflicting_pairs = [
            {'left_mirror', 'right_mirror'},
            {'left_headlight', 'right_headlight'},
            {'left_door', 'right_door'}
        ]

        for i in range(len(boxes)):
            if i in indices_to_drop: continue
            name_i = model_parts.names[cls_ids[i]]
            for j in range(i + 1, len(boxes)):
                if j in indices_to_drop: continue
                name_j = model_parts.names[cls_ids[j]]
                is_conflict = any({name_i, name_j} == pair for pair in conflicting_pairs)
                if is_conflict:
                    box1, box2 = boxes[i], boxes[j]
                    x_left, y_top = max(box1[0], box2[0]), max(box1[1], box2[1])
                    x_right, y_bottom = min(box1[2], box2[2]), min(box1[3], box2[3])
                    if x_right > x_left and y_bottom > y_top:
                        inter = (x_right - x_left) * (y_bottom - y_top)
                        area1, area2 = (box1[2]-box1[0])*(box1[3]-box1[1]), (box2[2]-box2[0])*(box2[3]-box2[1])
                        iou = inter / float(area1 + area2 - inter)
                        if iou > 0.3:
                            if confs[i] > confs[j]: indices_to_drop.add(j)
                            else: indices_to_drop.add(i)

    # --- PHASE 2: Mask Grouping & Unclassified Isolation ---
    grouped_masks = {}
    master_mask = np.zeros((h, w), dtype=np.uint8)
    if res_parts.masks is not None:
        valid_indices = [idx for idx in range(len(res_parts.boxes)) if idx not in indices_to_drop]
        for i in valid_indices:
            mask = res_parts.masks.data[i]
            part_name = model_parts.names[int(res_parts.boxes.cls[i])]
            m_bin = (cv2.resize(mask.cpu().numpy(), (w, h)) > 0.5).astype(np.uint8) * 255
            grouped_masks[part_name] = cv2.bitwise_or(grouped_masks.get(part_name, np.zeros((h, w), dtype=np.uint8)), m_bin)
            master_mask = cv2.bitwise_or(master_mask, m_bin)

    grouped_masks["Unclassified_Parts"] = cv2.bitwise_not(master_mask)

    # --- PHASE 3: Two-Pass Damage Scan & Unified List JSON ---
    # Unified list storage for the whole image
    final_report = {
        "part_name": [],
        "damage_type": [],
        "damage_ration": [],
        "confindence of model": []
    }

    for name, m_bin in grouped_masks.items():
        if not np.any(m_bin): continue
        seg_img = cv2.bitwise_and(orig_img, orig_img, mask=m_bin)

        # Two-Pass Scan
        res1 = model_damage.predict(source=seg_img, conf=0.25, verbose=False)[0]
        x, y, wb, hb = cv2.boundingRect(m_bin)
        res2 = model_damage.predict(source=seg_img[y:y+hb, x:x+wb], conf=0.25, verbose=False)[0] if wb > 10 and hb > 10 else None

        if (res1.boxes is not None and len(res1.boxes) > 0) or (res2 and res2.boxes is not None and len(res2.boxes) > 0):
            display_img = seg_img.copy()
            combined_dmg_mask = np.zeros((h, w), dtype=np.uint8)

            # Record damages
            current_part_dmg_types = []
            current_part_confs = []

            for res, is_crop in [(res1, False), (res2, True)]:
                if res is not None and res.boxes is not None:
                    for i in range(len(res.boxes)):
                        dmg_type = model_damage.names[int(res.boxes.cls[i])]
                        conf = f"{float(res.boxes.conf[i]):.2f}"
                        if dmg_type not in current_part_dmg_types:
                            current_part_dmg_types.append(dmg_type)
                            current_part_confs.append(conf)

                        # Handle visualization masks
                        if res.masks is not None:
                            dm = (cv2.resize(res.masks.data[i].cpu().numpy(), (wb, hb) if is_crop else (w, h)) > 0.5).astype(np.uint8)
                            if is_crop:
                                full_mask = np.zeros((h, w), dtype=np.uint8)
                                full_mask[y:y+hb, x:x+wb] = dm
                                combined_dmg_mask = cv2.bitwise_or(combined_dmg_mask, full_mask)
                            else:
                                combined_dmg_mask = cv2.bitwise_or(combined_dmg_mask, dm)

            # Ratio Calculation
            p_area, d_area = np.count_nonzero(m_bin), np.count_nonzero(combined_dmg_mask)
            ratio_str = f"{(d_area / p_area * 100):.2f}%" if p_area > 0 else "0.00%"

            # Final Logic: Add to global lists
            final_report["part_name"].append("unidentified_part" if name == "Unclassified_Parts" else name)
            final_report["damage_type"].extend(current_part_dmg_types)
            final_report["damage_ration"].append(ratio_str)
            final_report["confindence of model"].extend(current_part_confs)

            # Visuals
            if np.any(combined_dmg_mask):
                overlay = np.zeros_like(display_img)
                overlay[combined_dmg_mask == 1] = [0, 0, 255]
                display_img = cv2.addWeighted(display_img, 1.0, overlay, 0.6, 0)

            out_name = f"output_{name}.jpg"
            cv2.imwrite(out_name, display_img)
            display(Image(filename=out_name))

    print("\n" + "="*50 + "\nIMAGE SCAN JSON\n" + "="*50)
    print(json.dumps(final_report, indent=4))

# 2. RUN
uploaded = files.upload()
for filename in uploaded.keys():
    extract_and_assess_segments(filename, model_parts, model_damage)
    os.remove(filename)
```

```python
import os
import json
import torch
import numpy as np
import cv2
from ultralytics import YOLO
from google.colab import files
from IPython.display import display, Image

# 1. SETUP MODELS
parts_path = '/content/drive/MyDrive/CarParts_YOLO26/yolo26_carparts_best.pt'
if not os.path.exists(parts_path): parts_path = 'yolo26_carparts_best.pt'

damage_path = '/content/drive/MyDrive/CarParts_YOLO26/damage_detection_best.pt'
if not os.path.exists(damage_path): damage_path = 'damage_detection_best.pt'

model_parts = YOLO(parts_path)
model_damage = YOLO(damage_path)

def extract_and_assess_segments(filename, model_parts, model_damage):
    orig_img = cv2.imread(filename)
    h, w = orig_img.shape[:2]

    res_parts = model_parts.predict(source=orig_img, conf=0.25, verbose=False)[0]

    # --- PHASE 1: Cross-Class Overlap Filtering (Mirror Fix) ---
    indices_to_drop = set()
    if res_parts.boxes is not None and len(res_parts.boxes) > 1:
        boxes = res_parts.boxes.xyxy.cpu().numpy()
        confs = res_parts.boxes.conf.cpu().numpy()
        cls_ids = res_parts.boxes.cls.cpu().numpy().astype(int)

        conflicting_pairs = [{'left_mirror', 'right_mirror'}, {'left_headlight', 'right_headlight'}, {'left_door', 'right_door'}]

        for i in range(len(boxes)):
            if i in indices_to_drop: continue
            name_i = model_parts.names[cls_ids[i]]
            for j in range(i + 1, len(boxes)):
                if j in indices_to_drop: continue
                name_j = model_parts.names[cls_ids[j]]
                if any({name_i, name_j} == pair for pair in conflicting_pairs):
                    box1, box2 = boxes[i], boxes[j]
                    inter = max(0, min(box1[2], box2[2]) - max(box1[0], box2[0])) * max(0, min(box1[3], box2[3]) - max(box1[1], box2[1]))
                    iou = inter / float((box1[2]-box1[0])*(box1[3]-box1[1]) + (box2[2]-box2[0])*(box2[3]-box2[1]) - inter)
                    if iou > 0.3:
                        if confs[i] > confs[j]: indices_to_drop.add(j)
                        else: indices_to_drop.add(i)

    # --- PHASE 2: Grouping ---
    grouped_masks = {}
    master_mask = np.zeros((h, w), dtype=np.uint8)
    if res_parts.masks is not None:
        valid_indices = [idx for idx in range(len(res_parts.boxes)) if idx not in indices_to_drop]
        for i in valid_indices:
            part_name = model_parts.names[int(res_parts.boxes.cls[i])]
            m_bin = (cv2.resize(res_parts.masks.data[i].cpu().numpy(), (w, h)) > 0.5).astype(np.uint8) * 255
            grouped_masks[part_name] = cv2.bitwise_or(grouped_masks.get(part_name, np.zeros((h, w), dtype=np.uint8)), m_bin)
            master_mask = cv2.bitwise_or(master_mask, m_bin)

    grouped_masks["Unclassified_Parts"] = cv2.bitwise_not(master_mask)

    # --- PHASE 3: Cleaned JSON & Scan ---
    final_report = {"part_name": [], "damage_type": [], "damage_ration": [], "confindence of model": []}

    for name, m_bin in grouped_masks.items():
        if not np.any(m_bin): continue
        seg_img = cv2.bitwise_and(orig_img, orig_img, mask=m_bin)

        # 2-Pass Scan
        r1 = model_damage.predict(source=seg_img, conf=0.25, verbose=False)[0]
        x, y, wb, hb = cv2.boundingRect(m_bin)
        r2 = model_damage.predict(source=seg_img[y:y+hb, x:x+wb], conf=0.25, verbose=False)[0] if wb > 10 and hb > 10 else None

        # Check if any damage exists in this part
        d1 = r1.boxes is not None and len(r1.boxes) > 0
        d2 = r2 is not None and r2.boxes is not None and len(r2.boxes) > 0

        if d1 or d2:
            display_img = seg_img.copy()
            combined_dmg_mask = np.zeros((h, w), dtype=np.uint8)

            # Local sets to prevent duplication within the same part
            found_types = []
            found_confs = []

            for res, is_crop in [(r1, False), (r2, True)]:
                if res is not None and res.boxes is not None:
                    for i in range(len(res.boxes)):
                        t = model_damage.names[int(res.boxes.cls[i])]
                        c = f"{float(res.boxes.conf[i]):.2f}"
                        if t not in found_types:
                            found_types.append(t)
                            found_confs.append(c)

                        if res.masks is not None:
                            dm = (cv2.resize(res.masks.data[i].cpu().numpy(), (wb, hb) if is_crop else (w, h)) > 0.5).astype(np.uint8)
                            if is_crop:
                                full_dm = np.zeros((h, w), dtype=np.uint8)
                                full_dm[y:y+hb, x:x+wb] = dm
                                combined_dmg_mask = cv2.bitwise_or(combined_dmg_mask, full_dm)
                            else:
                                combined_dmg_mask = cv2.bitwise_or(combined_dmg_mask, dm)

            # Ratio Math (Fixed)
            p_area, d_area = np.count_nonzero(m_bin), np.count_nonzero(combined_dmg_mask)
            ratio_val = min(100.0, (d_area / p_area * 100)) if p_area > 0 else 0
            ratio_str = f"{ratio_val:.2f}%"

            # Update JSON List
            part_label = "unidentified_part" if name == "Unclassified_Parts" else name
            for i in range(len(found_types)):
                final_report["part_name"].append(part_label)
                final_report["damage_type"].append(found_types[i])
                final_report["damage_ration"].append(ratio_str)
                final_report["confindence of model"].append(found_confs[i])

            # Visualization
            if np.any(combined_dmg_mask):
                ov = np.zeros_like(display_img)
                ov[combined_dmg_mask == 1] = [0, 0, 255]
                display_img = cv2.addWeighted(display_img, 1.0, ov, 0.6, 0)

            out_n = f"output_{name}.jpg"
            cv2.imwrite(out_n, display_img)
            display(Image(filename=out_n))

    print("\n" + "="*20 + " CLEAN JSON REPORT " + "="*20)
    print(json.dumps(final_report, indent=4))

# 2. RUN
uploaded = files.upload()
for fn in uploaded.keys():
    extract_and_assess_segments(fn, model_parts, model_damage)
    os.remove(fn)
```

```python
import os
import json
import torch
import numpy as np
import cv2
from ultralytics import YOLO
from google.colab import files
from IPython.display import display, Image

# 1. SETUP MODELS
parts_path = '/content/drive/MyDrive/CarParts_YOLO26/yolo26_carparts_best.pt'
if not os.path.exists(parts_path): parts_path = 'yolo26_carparts_best.pt'

damage_path = '/content/drive/MyDrive/CarParts_YOLO26/damage_detection_best.pt'
if not os.path.exists(damage_path): damage_path = 'damage_detection_best.pt'

model_parts = YOLO(parts_path)
model_damage = YOLO(damage_path)

def analyze_damage_first(filename, model_parts, model_damage):
    orig_img = cv2.imread(filename)
    h, w = orig_img.shape[:2]

    # --- STEP 1: FIND DAMAGE LOCATIONS FIRST ---
    # We run a high-sensitivity scan to find every potential dent/scratch
    res_damage = model_damage.predict(source=orig_img, conf=0.20, verbose=False)[0]

    final_report = {
        "part_name": [],
        "damage_type": [],
        "damage_ration": [],
        "confindence of model": []
    }

    if res_damage.boxes is not None and len(res_damage.boxes) > 0:
        # --- STEP 2: SEGMENT THE CAR PARTS ---
        res_parts = model_parts.predict(source=orig_img, conf=0.25, verbose=False)[0]

        # Create a combined image for visualization
        display_img = orig_img.copy()

        # Loop through each detected damage instance
        for i, d_box in enumerate(res_damage.boxes):
            d_cls = int(d_box.cls[i]) if hasattr(d_box.cls, "__len__") else int(d_box.cls)
            dmg_type = model_damage.names[d_cls]
            dmg_conf = float(d_box.conf[i]) if hasattr(d_box.conf, "__len__") else float(d_box.conf)

            # Get damage coordinates
            dx1, dy1, dx2, dy2 = d_box.xyxy[0].cpu().numpy()

            # --- STEP 3: FIND WHICH SEGMENT CONTAINS THIS DAMAGE ---
            assigned_part = "unidentified_part"
            max_overlap = 0

            if res_parts.masks is not None:
                for j, p_mask in enumerate(res_parts.masks.data):
                    # Resize part mask to original image size
                    p_m_np = (cv2.resize(p_mask.cpu().numpy(), (w, h)) > 0.5).astype(np.uint8)

                    # Create a temporary mask for the damage box
                    d_mask_tmp = np.zeros((h, w), dtype=np.uint8)
                    cv2.rectangle(d_mask_tmp, (int(dx1), int(dy1)), (int(dx2), int(dy2)), 1, -1)

                    # Calculate intersection
                    overlap = np.sum(cv2.bitwise_and(p_m_np, d_mask_tmp))

                    if overlap > max_overlap:
                        max_overlap = overlap
                        assigned_part = model_parts.names[int(res_parts.boxes.cls[j])]

            # Calculate individual damage ratio relative to the assigned part
            # If part exists, ratio = (dmg_box_area / part_mask_area)
            # Otherwise, use full image area as fallback
            dmg_area = (dx2 - dx1) * (dy2 - dy1)

            # Update JSON
            final_report["part_name"].append(assigned_part)
            final_report["damage_type"].append(dmg_type)
            final_report["damage_ration"].append(f"{(dmg_area / (h*w) * 100):.2f}%")
            final_report["confindence of model"].append(f"{dmg_conf:.2f}")

            # Visualization: Draw Damage Box and Label
            cv2.rectangle(display_img, (int(dx1), int(dy1)), (int(dx2), int(dy2)), (0, 0, 255), 3)
            cv2.putText(display_img, f"{assigned_part}: {dmg_type}", (int(dx1), int(dy1)-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

        # Show Output
        cv2.imwrite('identified_damage.jpg', display_img)
        display(Image(filename='identified_damage.jpg'))
    else:
        print("✅ No damage found to identify.")

    print("\n" + "="*20 + " TARGETED DAMAGE REPORT " + "="*20)
    print(json.dumps(final_report, indent=4))

# 2. RUN
uploaded = files.upload()
for fn in uploaded.keys():
    analyze_damage_first(fn, model_parts, model_damage)
    os.remove(fn)
```
